```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
total_schools = len(school_data['School ID'])
total_students= len(student_data['Student ID'])
total_budget = school_data["budget"].sum()

```

```python
avg_math = student_data['math_score'].mean()
avg_reading = student_data['reading_score'].mean()

```

```python
passing_math = len(student_data[student_data['math_score'] >= 70])
per_passingmath = (passing_math/total_students)*100
                             
                            
                             
```

```python
passing_reading = len(student_data[student_data['reading_score'] >= 70])
per_passingreading = (passing_reading/total_students)*100

```

```python
passing_both = len(student_data[(student_data['reading_score'] >= 70) &
                    (student_data['math_score'] >= 70)])
per_passingboth = (passing_both/total_students)*100
```

```python
district_summary_df = pd.DataFrame([
    {"Total Schools": total_schools, "Total Students" : '{:,}'.format(total_students), "Total Budget": '${:,}'.format(total_budget), "Average Math Score": avg_math ,"Average Reading Score": avg_reading, "% Passing Math": per_passingmath, "% Passing Reading": per_passingreading, "% Overall Passing": per_passingboth}
])
district_summary_df
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
summary_byschool_df = school_data_complete.groupby(["school_name"])
# school_name = summary_byschool_df['school_name'].unique()
total_budget_summary = list(summary_byschool_df["budget"].mean())
# school_type = list(summary_byschool_df['type'].unique())
total_students_summary = list(summary_byschool_df['Student ID'].count())
perstudent_budget = [i/j for i,j in zip(total_budget_summary, total_students_summary)]
avg_math_summary = list(summary_byschool_df['math_score'].mean())
avg_reading_summary = list(summary_byschool_df['reading_score'].mean())
```

```python
summary_byschool_df = school_data_complete[school_data_complete['math_score'] >= 70].groupby(["school_name"])
per_passingmath_summary = [(i/j)*100 for i,j in zip(summary_byschool_df['math_score'].count(), total_students_summary)]

summary_byschool_df = school_data_complete[school_data_complete['reading_score'] >= 70].groupby(["school_name"])
per_passingreading_summary = [(i/j)*100 for i,j in zip(summary_byschool_df['reading_score'].count(), total_students_summary)]

# passing_both = school_data_complete[school_data_complete['math_score'] >= 70] & ([school_data_complete['reading_score'] >= 70].groupby(["school_name"]))
# # passing_both = len(student_data[(student_data['reading_score'] >= 70) & (student_data['math_score'] >= 70)])

school_summary_df = pd.DataFrame({"School Names":school_data["school_name"],
                                  "School Type":school_data["type"],
                                  "Total Students":total_students_summary,
                                  "Total School Budget":total_budget_summary,
                                  "Per Student Budget": perstudent_budget,
                                  "Average Math Score":avg_math_summary,
                                  "Average Reading Score":avg_reading_summary,
                                  "% Passing Math":per_passingmath_summary,
                                  "% Passing Reading":per_passingreading_summary
                                 })
school_summary_df.reset_index(drop=True)
school_summary_df.set_index('School Names')
school_summary_df.rename_axis(None)
```

## Top Performing Schools (By % Overall Passing)

```python
top_schools = school_summary_df.sort_values("% Passing Math", ascending = False).head(5)
top_schools.rename_axis(None)
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
ninth_grade = school_data_complete[school_data_complete['grade']=="9th"].groupby('school_name').mean()['math_score']
tenth_grade = school_data_complete[school_data_complete['grade']=="10th"].groupby('school_name').mean()['math_score']
eleventh_grade = school_data_complete[school_data_complete['grade']=="11th"].groupby('school_name').mean()['math_score']
twelfth_grade = school_data_complete[school_data_complete['grade']=="12th"].groupby('school_name').mean()['math_score']

math_grade_df = pd.DataFrame({"9th": ninth_grade, '10th':tenth_grade,'11th':eleventh_grade,'12th':twelfth_grade})
math_grade_df.rename_axis(None)


```

## Reading Score by Grade 

* Perform the same operations as above for reading scores

```python
ninth_grade = school_data_complete[school_data_complete['grade']=="9th"].groupby('school_name').mean()['reading_score']
tenth_grade = school_data_complete[school_data_complete['grade']=="10th"].groupby('school_name').mean()['reading_score']
eleventh_grade = school_data_complete[school_data_complete['grade']=="11th"].groupby('school_name').mean()['reading_score']
twelfth_grade = school_data_complete[school_data_complete['grade']=="12th"].groupby('school_name').mean()['reading_score']

reading_grade_df = pd.DataFrame({"9th": ninth_grade, '10th':tenth_grade,'11th':eleventh_grade,'12th':twelfth_grade})
reading_grade_df.rename_axis(None)
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
spending_bins = [0, 584, 629,644,675]
range_names = ["<$584", "$585-629", "$630-644", "$645-675"]

school_summary_df['Spending Ranges (Per Student)'] = pd.cut(school_summary_df["Per Student Budget"], spending_bins, labels=range_names)

school_spend_group = school_summary_df.groupby('Spending Ranges (Per Student)').mean()
school_spend_group_df = school_spend_group[[
                                    "Average Math Score",
                                    "Average Reading Score",
                                    "% Passing Math",
                                    "% Passing Reading"]] 

school_spend_group_df
```

## Scores by School Size

* Perform the same operations as above, based on school size.

```python
size_bins = [0, 1000, 2000, 5000]
range_names = ["<1000", "1000-2000", "2000-5000"]

school_summary_df['School Size'] = pd.cut(school_summary_df["Total Students"], size_bins, labels=range_names)

school_size_group = school_summary_df.groupby('School Size').mean()
school_size_group_df = school_size_group[[
                                    "Average Math Score",
                                    "Average Reading Score",
                                    "% Passing Math",
                                    "% Passing Reading"]] 

school_size_group_df
```

## Scores by School Type

* Perform the same operations as above, based on school type

```python

school_type_group = school_summary_df.groupby("School Type").mean()
school_type_group_df = school_type_group[[
                                    "Average Math Score",
                                    "Average Reading Score",
                                    "% Passing Math",
                                    "% Passing Reading"]] 

school_type_group_df

```
